IMPORT LIBRARY

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# READ CSV
csv = pd.read_csv('https://raw.githubusercontent.com/adenafil/imk_data_kuisoner/main/MyAnimeBoxd.csv')
csv
```

DATA NAMA DAN GENDER #1

```python
namaDanGender = pd.concat([csv['Nama'], csv['Jenis Kelamin']], axis = 1)
namaDanGender = namaDanGender.drop(namaDanGender.index[0])
# Replace NaN values in 'Jenis Kelamin' column with 'Pria'
namaDanGender['Jenis Kelamin'] = namaDanGender['Jenis Kelamin'].fillna('Pria')

namaDanGender
```

Visualisasi Data Gender #1

```python
# Hitung Data
jumlah_pria = namaDanGender['Jenis Kelamin'].value_counts()['Pria']
jumlah_wanita = namaDanGender['Jenis Kelamin'].value_counts()['Wanita']

# Visualize the data
labels = ['Pria', 'Wanita']
sizes = [jumlah_pria, jumlah_wanita]
colors = ['#ff9999', '#66b3ff']
explode = (0.1, 0) # slice between two genders

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Jumlah Pria Dan Wanita')
plt.show()
```

Apakah menurutmu penting untuk melacak apa yang sudah kita tonton agar tidak lupa? Jika pernah ceritakan Pengalaman anda # 4

```python
isTrackingImportant = pd.concat([csv['Nama'], csv['\nApakah menurutmu penting untuk melacak apa yang sudah kita tonton agar tidak lupa? Jika pernah ceritakan Pengalaman anda']], axis = 1)
isTrackingImportant = isTrackingImportant.drop(isTrackingImportant.index[0])
isTrackingImportant
```

```python
def normalize_response(response):
    if isinstance(response, str):  # Check if response is a string
        # List of keywords
        keywords = ['iya', 'ya', 'yes', 'yoi', 'ok', 'sudah', 'pernah', 'perlu', 'ketiduran', 'cukup', 'penting', 'mungkin']

        # Check if any keyword is present in the response
        if any(keyword in response.lower() for keyword in keywords):
            # print(response, 'IYA PERNAH') #ACTIVATE ONLY FOR DEBUGGERS 😎
            return 'IYA PERNAH'
        else:
            # print(response, 'TIDAK')  #ACTIVATE ONLY FOR DEBUGGERS 😎
            return 'Tidak'
    else:
        # print(response, 'TIDAK')  #ACTIVATE ONLY FOR DEBUGGERS 😎
        return 'Tidak'  # If response is NaN or not a string, return 'Tidak'

#BELOW TEKS FOR DEBUGGERS 😎
# Apply normalization function
# isTrackingImportant['\nApakah menurutmu penting untuk melacak apa yang sudah kita tonton agar tidak lupa? Jika pernah ceritakan Pengalaman anda'] = isTrackingImportant['\nApakah menurutmu penting untuk melacak apa yang sudah kita tonton agar tidak lupa? Jika pernah ceritakan Pengalaman anda'].apply(normalize_response)

# # Hitung Data
# value_counts = isTrackingImportant['\nApakah menurutmu penting untuk melacak apa yang sudah kita tonton agar tidak lupa? Jika pernah ceritakan Pengalaman anda'].value_counts()

# # Mendapatkan jumlah kemunculan nilai tertentu
# penting_count = value_counts.get('IYA PERNAH', 0)
# tidak_penting_count = value_counts.get('Tidak', 0)

# print("Jumlah 'IYA PERNAH':", penting_count)
# print("Jumlah 'Tidak':", tidak_penting_count)
```

Visualisasi Data #4

```python
# Hitung Data
data_iya_or_no = isTrackingImportant['\nApakah menurutmu penting untuk melacak apa yang sudah kita tonton agar tidak lupa? Jika pernah ceritakan Pengalaman anda'].apply(normalize_response)
iya = data_iya_or_no.value_counts()['IYA PERNAH']
tidak = data_iya_or_no.value_counts()['Tidak']

# Visualize the data
labels = ['Yang Setuju Fitur Tracking Penting', 'Yang Tidak Setuju Fitur Tracking Penting']
sizes = [iya, tidak]
colors = ['#ff9999', '#66b3ff']
explode = (0.1, 0) # slice between two genders

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Apakah Fitur Tracking Penting ?')
plt.show()
```

Apa tanggapan anda terhadap fitur seperti list anime, save your watch list ? #7

Visual Data #7

```python
def messi_parser(response):
    if isinstance(response, str):  # Check if response is a string
        # List of keywords
        keywords = ['membantu', 'sangat', 'penting', 'bisa', 'bagus', 'cukup', 'mayan', 'butuh', 'boleh', 'mungkin', 'berguna', 'dibutuhkan', 'jadi', 'masuk', 'bookmark']

        # Check if any keyword is present in the response
        if any(keyword in response.lower() for keyword in keywords):
            # print(response, 'SUKA') #ACTIVATE ONLY FOR DEBUGGERS 😎
            return 'SUKA'
        else:
            # print(response, 'TIDAK')  #ACTIVATE ONLY FOR DEBUGGERS 😎
            return 'Tidak'
    else:
        # print(response, 'TIDAK')  #ACTIVATE ONLY FOR DEBUGGERS 😎
        return 'Tidak'  # If response is NaN or not a string, return 'Tidak'
```

```python
# Hitung Data
data_iya_or_no = tanggapan_fitur['Apa tanggapan anda terhadap fitur seperti list anime, save your watch list ? '].apply(messi_parser)
suka = data_iya_or_no.value_counts()['SUKA']
tidak_suka = data_iya_or_no.value_counts()['Tidak']

# Visualize the data
labels = ['Yang Setuju Fitur List Anime', 'Yang Tidak Setuju Fitur List Anime']
sizes = [suka, tidak_suka]
colors = ['#ff9999', '#66b3ff']
explode = (0.1, 0) # slice between two genders

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Charts(apakah suka fitur list anime)')
plt.show()
```
